```python
hsv = cv2.cvtColor(A, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv)
```

```python
mask = cv2.inRange(hsv, (0, 75, 40), (180, 255, 255))
```

```python
mask_3d = np.repeat(mask[:, :, np.newaxis], 3, axis=2)
```

```python
blurred_frame = cv2.GaussianBlur(A, (25, 25), 0)
```

```python
frame = np.where(mask_3d == (255, 255, 255), A, blurred_frame)
```

```python
plt.imshow(frame)
```
